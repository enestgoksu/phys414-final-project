```python
import pandas as pd

df = pd.read_csv("../data/white_dwarf_data.csv")
df.head()
```

```python
df.columns
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants (CGS)
G = 6.67430e-8        # cm^3 g^-1 s^-2
M_sun = 1.98847e33   # g
R_earth = 6.371e8    # cm

# Extract data
M = df["mass"].values          # solar masses
logg = df["logg"].values      # log10(g) in CGS

# Convert to CGS
g = 10**logg
M_cgs = M * M_sun

# Radius from Newtonian gravity
R_cgs = np.sqrt(G * M_cgs / g)
R_Re = R_cgs / R_earth   # Earth radii

# Plot M-R relation
plt.figure()
plt.scatter(R_Re, M, s=15)
plt.xlabel("R (Earth radii)")
plt.ylabel("M (Solar masses)")
plt.title("White Dwarf Massâ€“Radius Data")
plt.show()

# Quick sanity checks
print("R range (Earth radii):", R_Re.min(), "to", R_Re.max())
print("M range (Solar masses):", M.min(), "to", M.max())
```
